# Initializing Project. 🚀

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('transaction_records.csv')
pd.read_csv('transaction_records.csv')
```

# Using Numpy 👷‍♂️ for numerical data.

```python
mean_amount = np.mean(df["Amount"])
print(mean_amount)
```

```python
std_balance = np.std(df["AccountBalance"])
print(std_balance)
```

```python
max_amount = np.max(df["Amount"])
print(max_amount)
```

```python
min_balance = np.min(df["AccountBalance"])
print(min_balance)
```

```python
variance_balance = np.var(df["AccountBalance"])
print(variance_balance)
```

```python
median_amount = np.median(df["Amount"])
print(median_amount)
```

# Using **Pandas** 🐼 for Manupilation.

```python
df["TransactionPercentage"] = (df["Amount"] / df["AccountBalance"]) * 100
df
```

```python
# using pandas
shape = df.shape
print("Shape of the Dataframe:", shape)
```

```python
first_five_columns = df.iloc[:, :5]
display(first_five_columns)
```

```python
# transactions greater then mean
high_transactions = df[df["Amount"] > mean_amount]
print(high_transactions)
```

```python
print(df.info())
```

```python
top_five_transactions = df['Amount'].nlargest(5)
print("Top Five Transactions Values:\n", top_five_transactions)
```

```python
customer_total = df.groupby("CustomerID")["Amount"].sum().reset_index()
print(customer_total)
```

```python
average_amount = df['Amount'].mean()
print(f"Average Amount transferred: {average_amount}")
```

```python
max_amount = df['Amount'].max()
print(f"Maximum Amount transffered: {max_amount}")
```
